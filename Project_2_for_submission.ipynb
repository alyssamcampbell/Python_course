```python
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 24
neighbors_settings = range(1, 24)
for n_neighbors in neighbors_settings:
 # build the model
 knn = KNeighborsClassifier(n_neighbors=n_neighbors)
 knn.fit(X_train, y_train)
 # record training set accuracy
 training_accuracy.append(knn.score(X_train, y_train))
 # record generalization accuracy
 test_accuracy.append(knn.score(X_test, y_test))
 
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
#MANHATTAN DISTANCE VERSUS DEFAULT MINKOWSKI/EUCLIDEAN DISTANCE
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 24
neighbors_settings = range(1, 24)
for n_neighbors in neighbors_settings:
 # build the model
 knn = KNeighborsClassifier(n_neighbors=n_neighbors,metric='manhattan')
 knn.fit(X_train, y_train)
 # record training set accuracy
 training_accuracy.append(knn.score(X_train, y_train))
 # record generalization accuracy
 test_accuracy.append(knn.score(X_test, y_test))
 
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
# WEIGHTED WITH INVERSE OF DISTANCE (VS DEFAULT UNIFORM WEIGHTING)
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 24
neighbors_settings = range(1, 24)

for n_neighbors in neighbors_settings:
    # build the model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors,weights='distance')
    knn.fit(X_train, y_train)
    # record training set accuracy
    training_accuracy.append(knn.score(X_train, y_train))
    # record generalization accuracy
    test_accuracy.append(knn.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
#Experimenting with changing the algorithm to compute nearest neighbors (ball_tree, kd_tree ,brute,'auto)
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 24
neighbors_settings = range(1, 24)

for n_neighbors in neighbors_settings:
    # build the model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors,algorithm="brute")
    knn.fit(X_train, y_train)
    # record training set accuracy
    training_accuracy.append(knn.score(X_train, y_train))
    # record generalization accuracy
    test_accuracy.append(knn.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
#TESTING A SERIES OF DIFFERENT DEPTHS - MAX NUMBER OF LEVEL IN EACH DECISION TREE 
training_accuracy = []
test_accuracy = []
# try max_depth from 1 to 10
ensemble_settings = range(1, 10)
for max_depth in ensemble_settings:
 # build the model
 model = RandomForestClassifier(max_depth=max_depth)
 model.fit(X_train, y_train)
 # record training set accuracy
 training_accuracy.append(model.score(X_train, y_train))
 # record generalization accuracy
 test_accuracy.append(model.score(X_test, y_test))
 
plt.plot(ensemble_settings, training_accuracy, label="training accuracy")
plt.plot(ensemble_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Max_depth")
plt.legend()
```

```python
#TESTING A SERIES OF DIFFERENT DEPTHS - MAX FEATURES CONSIDERED FOR SPLITTING A NODE
training_accuracy = []
test_accuracy = []
# try max_features from 1 to 10
ensemble_settings = range(1, 10)
for max_features in ensemble_settings:
 # build the model
 model = RandomForestClassifier(max_features=max_features)
 model.fit(X_train, y_train)
 # record training set accuracy
 training_accuracy.append(model.score(X_train, y_train))
 # record generalization accuracy
 test_accuracy.append(model.score(X_test, y_test))
 
plt.plot(ensemble_settings, training_accuracy, label="training accuracy")
plt.plot(ensemble_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Max_features")
plt.legend()
```
